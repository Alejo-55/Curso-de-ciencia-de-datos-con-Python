## Seleccionar filas

### loc[] vs iloc[]

La función loc[] permite acceder a los elementos a partir de su etiqueta, que normalmente es el número de fila.

La función iloc[] permite acceder a los elementos a traves de su posición y no de su etiqueta.

```python
# El comando loc[] selecciona filas.

df.loc[[0,1,2]]
```

```python
# El comando iloc[] también selecciona filas
df.iloc[0:3]
```

# Groupby

El  groupby sirve para separar/filtar datos. Por ejemplo, comparar resultados entre el sexo masculino y femenino. Ejemplo:

```python
# Ejemplo, promedio de edad entre hombres y mujeres (primera columna de la tabla de resultados).
df.groupby("sex").mean()
```

```python
## Personas que tienen problemas cardíacos, con el count.
# Dice las personas no tienen problemas cardíacos (el cero), y las que sí tienen problemas cardíacos( el uno)
df.groupby("target").count()
```
